```python
import numpy as np
import pandas as pd
from scipy import stats
import seaborn as sns
data= pd.read_csv('./dataR2.csv')
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Veriyi okuma işlemi
data = pd.read_csv('./dataR2.csv')

# Veri dağılımı grafikleri
for column in data.columns:
    plt.figure(figsize=(4, 2))
    sns.histplot(data[column], bins=20, kde=True)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
data.isnull().sum()
```

```python
data.duplicated()
```

```python
data=data.drop_duplicates()
data.info()
```

```python
for column in data.columns:
    data[column] = (data[column] - data[column].min()) / (data[column].max() - data[column].min())
data.head()
```

```python
y=data['Classification']
x=data.drop(['Classification'],axis=1)
```

```python
import matplotlib.pyplot as plt

# Count the occurrences of each class in 'y'
class_counts = y.value_counts()

# Create a bar chart
plt.figure(figsize=(4, 3))
class_counts.plot(kind='bar', color='skyblue')
plt.title('Class Distribution')
plt.xlabel('Class')
plt.ylabel('Count')
plt.xticks(rotation=0)  # Rotate x-axis labels if needed
plt.show()
```

```python
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# Split the data
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.1, stratify=y)

# Print the sizes of the training and test sets
print("Training Set Size:", xtrain.shape)
print("Test Set Size:", xtest.shape)
print("Training Label Set Size:", ytrain.shape)
print("Test Label Set Size:", ytest.shape)

# Calculate the proportions of the training and test sets
train_proportion = len(xtrain) / (len(xtrain) + len(xtest))
test_proportion = len(xtest) / (len(xtrain) + len(xtest))

# Create a pie chart
plt.figure(figsize=(4, 4))  # Set the desired dimensions (4 inches wide, 3 inches tall)
labels = 'Training Set', 'Test Set'
sizes = [train_proportion, test_proportion]
colors = ['lightcoral', 'lightskyblue']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Data Split Proportion')
plt.show()

```

```python
from sklearn.svm import SVC
svm_model= SVC(kernel='poly',gamma=8)
svm_model.fit(xtrain,ytrain)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import pandas as pd
import seaborn as sns

# Eğitim verisi üzerinde tahminler ve değerlendirme
predictions_train = svm_model.predict(xtrain)
percentage=svm_model.score(xtrain,ytrain)
confusion_matrix_train = confusion_matrix(ytrain, predictions_train)
classification_report_train = classification_report(ytrain, predictions_train, output_dict=True)

# Eğitim verisi sınıflandırma raporunu bir tabloya dökme
classification_df_train = pd.DataFrame(classification_report_train).transpose()

plt.figure(figsize=(4, 2))
sns.heatmap(confusion_matrix_train, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Training Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

print("\nTraining Classification Report:")
print(classification_df_train)
print("\nTraining Accuracy:", percentage*100)

# Test verisi üzerinde tahminler ve değerlendirme
predictions_test = svm_model.predict(xtest)
percentage1=svm_model.score(xtest,ytest)
confusion_matrix_test = confusion_matrix(ytest, predictions_test)
classification_report_test = classification_report(ytest, predictions_test, output_dict=True)

# Test verisi karmaşıklık matrisini ve sınıflandırma raporunu bir tabloya dökme
classification_df_test = pd.DataFrame(classification_report_test).transpose()

plt.figure(figsize=(4, 2))
sns.heatmap(confusion_matrix_test, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Validation Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# Tabloyu yazdırma
print("\nValidation Classification Report:")
print(classification_df_test*100)
print("\nValidation Accuracy:", percentage1*100)
```

```python
print(data['Classification'].value_counts())
df_class_0 = data[data['Classification'] == 0]
df_class_1 = data[data['Classification'] == 1]
```

```python
df_class_1_over = df_class_1.sample(250, replace=True)
df_class_0_over = df_class_0.sample(250, replace=True)
df_test_over = pd.concat([df_class_0_over, df_class_1_over], axis=0)
df_test_over.info()
```

```python
y1=df_test_over['Classification']
df_test_over=df_test_over.drop(['Classification'],axis=1)
X1=df_test_over
```

```python
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# Verileri bölmek
X1_s_train, X1_s_test, y1_s_train, y1_s_test = train_test_split(X1, y1, test_size=0.25, random_state=0, shuffle=True, stratify=y1)

# Eğitim ve test setlerinin boyutlarını yazdırma
print('Training data shape is :{}.'.format(X1_s_train.shape))
print('Training label shape is :{}.'.format(y1_s_train.shape))
print('Testing data shape is :{}.'.format(X1_s_test.shape))
print('Testing label shape is :{}.'.format(y1_s_test.shape))

# Veri bölünmesini gösteren bir pasta grafiği oluşturma
plt.figure(figsize=(4, 4.5))  # Grafiğin boyutunu ayarlayın
labels = 'Training Set', 'Test Set'
sizes = [len(X1_s_train), len(X1_s_test)]
colors = ['lightcoral', 'lightskyblue']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Dairesel bir pasta grafiği oluşturur
plt.title('Data Split Proportion')
plt.show()
```

```python
from sklearn.svm import SVC
svc_s_model = SVC(kernel='rbf',gamma=8)
svc_s_model.fit(X1_s_train, y1_s_train)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Eğitim verisi üzerinde tahminler ve değerlendirme
predictions_train = svc_s_model.predict(X1_s_train)
percentage=svc_s_model.score(X1_s_train,y1_s_train)
confusion_matrix_train = confusion_matrix(y1_s_train, predictions_train)
classification_report_train = classification_report(y1_s_train, predictions_train, target_names=['0.0', '1.0'], output_dict=True)

# Eğitim verisi  sınıflandırma raporunu bir tabloya dökme
classification_df_train = pd.DataFrame(classification_report_train).transpose()

# Eğitim verisi karmaşıklık matrisi görselleştirme
plt.figure(figsize=(4, 2))
sns.heatmap(confusion_matrix_train, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Training Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

print("\nTraining Classification Report:")
print(classification_df_train)
print("\nTraining Accuracy:", percentage*100)


# Test verisi üzerinde tahminler ve değerlendirme
predictions_test = svc_s_model.predict(X1_s_test)
percentage1=svc_s_model.score(X1_s_test,y1_s_test)
confusion_matrix_test = confusion_matrix(y1_s_test, predictions_test)
classification_report_test = classification_report(y1_s_test, predictions_test, target_names=['0.0', '1.0'], output_dict=True)

classification_df_test = pd.DataFrame(classification_report_test).transpose()


# Test verisi karmaşıklık matrisi görselleştirme
plt.figure(figsize=(4, 2))
sns.heatmap(confusion_matrix_test, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Validation Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()


print("\nValidation Classification Report:")
print(classification_df_test)
print("\nValidation Accuracy:", percentage1*100)
```

```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Dropout
dl_model = Sequential()

dl_model.add(Dense(256,  activation = 'relu' ,input_shape=([9]))) #input layer
dl_model.add(Dense(512,  activation = 'relu'))
dl_model.add(Dense(1,activation = 'sigmoid'))
dl_model.summary()
dl_model.compile(optimizer = 'adam' , loss = 'binary_crossentropy' ,metrics = ['accuracy','Precision','Recall','AUC'])
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint
filepath='best_model.h5'
checkpoint = ModelCheckpoint(filepath, monitor='val_accuracy', verbose=1, save_best_only=True, mode='max')
num_epochs = 20
history = dl_model.fit(X1_s_train ,
                    y1_s_train ,
                    epochs= num_epochs ,
                    steps_per_epoch=50,
                    validation_data=(X1_s_test ,y1_s_test),callbacks=checkpoint)
```

```python
#model performans grafiği
plt.figure(figsize=(6, 4))
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Training Metrics')
plt.xlabel('Epoch')
plt.ylabel('Metric Value')
plt.legend()
plt.show()
```
